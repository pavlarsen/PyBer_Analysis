```python
# Add error bars to a line chart
#standard deviation, standard error, confidence intervals, min, max values. Matlab uses errorbar()->plt.errorbar(x,y,yerr=<value>)
```

```python
%matplotlib inline
```

```python
# import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x_axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y_axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Calculate the stadnard deviation of the y_axis values.

stdev= statistics.stdev(y_axis)
stdev
```

```python
# Add the x_axis, y_axis and yerr=stdev to the errorbar function
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# Add a cap to the error bar that's equal to a scalar vlaue. Use capsize=
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# use the object oriented method
# Set the x_axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y_axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize = 3)
plt.show()
```

```python
# Add Error bars to Bar Charts Matlab way
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add Error bars to Bar Charts Object Oriented way
fig, ax=plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Change the Mayor Ticks using xticks()
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0,51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Change the Mayor ticks using object oriented xticks()

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51, step=5.0))
plt.show()
```

```python
# Add minor ticks with set_minor_locator
from matplotlib.ticker import MultipleLocator
#increase the size of the plot figure
fig, ax = plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51, step=5.0))
#Create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
