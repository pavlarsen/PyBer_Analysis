```python
#Create a line chart using the Matlab Method
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US Dollars accumulatd for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.24, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create a Line Chart using the Object-Oriented interface
#we use ax.plot() instead of plt.plot(). To create a figure enter: fix, ax = plt.subplots()
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#substitute fix, ax = plt.subplots() with other method
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#to create a quick graph using object oriented follow:
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#to change figure level attributes (axis labels, title or legend) or save the figure as an image, use "fig=plt.figure()"
```

```python
#use plt.show(). Should be only used once per session or it can display improperly
plt.plot(x_axis,y_axis)
plt.show()
```

```python
#methods used for annotate charts using Matlab method:
#plt.figure(figsize=(w, h)) -> change the size of the figure in pixels, added on the first line of the script.
#plt.plot(x,y, label='line') -> add a label that will be added to the legend.
#plt.xlim(min, max)-> Set the min and max range of the x-axis
#plt.ylim(min, max)-> Set the min and max range of the y-axis
#plt.xlabel('x label')-> add a label to the x-axis
#plt.ylabel ('y label')-> add a label to the y-axis
#plt.title('title')-> add a title
#plt.legend()-> add a legend
#plt.grid()-> add a grid to the chart
#plt.savefig("add a path and figure extension")-> save the figure with the given extension. Added at the end of the script.
```

```python
#Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color = "blue", linewidth=2, label='Boston')

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

#Create a title.
plt.title("PyBer Fare by Month")

#Add a grid
plt.grid()

#Add the legend
plt.legend()
```

```python
#methods used for annotate charts using Object-Oriented method:
#fig, ax= plt.subplots(figsize=(w, h)) -> change the size of the figure in pixels, added on the subplots() function.
#ax.plot(x,y, label='line') -> add a label that will be added to the legend.
#ax.set_xlim(min, max)-> Set the min and max range of the x-axis
#ax.set_ylim(min, max)-> Set the min and max range of the y-axis
#ax.set_xlabel('x label')-> add a label to the x-axis
#ax.set_ylabel ('y label')-> add a label to the y-axis
#ax.set_title('title')-> add a title
#ax.legend()-> add a legend
#ax.grid()-> add a grid to the chart
#** plt.savefig("add a path and figure extension")-> save the figure with the given extension. Added at the end of the script.
```

```python
fig, ax = plt.subplots()

#Create the plot and add a label for the legend.
ax.plot(x_axis, y_axis, marker="D", color = "green", linewidth=2, label='Boston')

#Create labels for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

#Set the y limit between 0 and 45
ax.set_ylim(0, 45)

#Create a title.
ax.set_title("PyBer Fare by Month")

#Add a grid
ax.grid()

#Add the legend
ax.legend()
```

```python
# Create Scatter Plots using Matlab method which has 2 methods: plt.plot() & plt.scatter()
#Using plt.plot()
# Set the x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US Dollars accumulatd for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.24, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# for plt.plot()method, when graphin, add a 'o'
plt.plot(x_axis, y_axis, 'o')
```

```python
# use plt.scatter()
plt.scatter(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(y_axis, x_axis, 'o', color="red", label = 'Chicago')
plt.gca().invert_yaxis()

# Create labels for the x and y axes
plt.xlabel("Fare($)")
plt.ylabel("Date")

#Set the x limit between 0 and 45
plt.xlim(0, 50)

#Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create a bubble chart using the matlab method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Some markers are too small, let's make them bigger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# you can use "list comprehension" to replace many for and while loops -> new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# Create a scatter plot using the Object Oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart using the object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax=plt.subplots()
# Create the plot
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color="skyblue", alpha=0.8, label = 'Boston', linewidth=2, edgecolor='k')
plt.gca().invert_yaxis()

# Create labels for the x and y axes
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

#Set the x limit between 0 and 45
ax.set_xlim(0, 50)

#Create a title
ax.set_title("PyBer Fare by Month")

# Add the legend
ax.legend()
```

```python
# Create Pie Charts
#Create Pie Chart using Matlab method
# Set the x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US Dollars accumulatd for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.24, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create pie chart using plt.pie()
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# use "explode" parameter to offset indicated wedge
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Change color for each wedge

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8,8))

plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Create a pie chart using Object Oriented Interface

# Set the x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US Dollars accumulatd for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.24, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Use object oriented method

fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
color_new = ["slateblue", "magenta", "pink", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values_new= (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis, explode=explode_values_new, colors=color_new, labels=x_axis, autopct='%.1f%%', shadow='True', startangle=95, counterclock=False)

plt.show()
```
